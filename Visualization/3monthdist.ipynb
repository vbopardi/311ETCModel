```python
import pandas as pd
import numpy as np 
import altair as alt
from altair import Chart, X, Y, Color, Scale
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
```

```python
df = pd.read_csv('../../fservice.csv')
```

```python
df['Just Date'] = df['Just Date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
```

```python
def histograms(df, start_date, end_date, binrange, axes, ticks):
    plt.hist(df[(df['Just Date'] >= start_date) & (df['Just Date'] <= end_date)]['ElapsedDays'], 
             bins = binrange, 
            edgecolor = 'Black')
    plt.title('Between ' + str(start_date)[:10] + ' and ' + str(end_date)[:10])
    plt.xlabel('ElapsedDays')
    plt.axis(axes)
    plt.xticks(ticks)
    plt.show()
    
    while end_date <= datetime.strptime('12/31/2019','%m/%d/%Y'):    
        start_date = start_date + relativedelta(months = +3)
        end_date = end_date + relativedelta(months = +3)
        return histograms(df, start_date, end_date, binrange, axes, ticks)
```

# 3-Month Distributions Where ElapsedDays Is Less Than 24

```python
histograms(df,
          datetime.strptime('1/1/2017','%m/%d/%Y'), 
          datetime.strptime('4/1/2017','%m/%d/%Y'),
          range(0,25,3),
          (0, 24, 0, 215000),
          np.arange(0, 25, 3))
```

# 3-Month Distributions Where ElapsedDays Is Less Than 9

```python
histograms(df,
          datetime.strptime('1/1/2017','%m/%d/%Y'), 
          datetime.strptime('4/1/2017','%m/%d/%Y'),
          range(0,10,1),
          (0, 9, 0, 92000),
          np.arange(0, 10, 1))
```
